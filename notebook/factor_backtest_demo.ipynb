# Demo for factor backtesting

```python
from lwpackage.lwfactors import BackTester
from lwpackage.lwmodels import LightGBM
```

## 1. stock factor

```python
bt = BackTester(fc_number_list=[1, 2, 3, 4], 
                instrument_type='stock',
                start_time='20160101', 
                end_time='20200101', 
                fc_freq = '1d',
                risk_free_rate=True,
                cost=0.00025,
                quantile=0.2,
                ir=True,
                longshort_instrument_number=5,
                yearly_performance=True,
                )
```

```python
bt.backtest()
```

```python
bt.bbt[1].data
```

### 1. general performance

```python
bt.performance
```

### 2. plot 

```python
bt.plot_nav(3)
```

```python
bt.plot_longshort(3)
```

# 2. futures factor

# 3. model-generated factor

use LightGBM as example 

```python
lgbm = LightGBM(fc_number_list=[1, 2, 3, 4, 5], 
                search_type = 'random',
                search_number = 10,
                early_stopping_rounds = 10,
                eval_metric = 'ic',
                start_time='20161231', 
                end_time='20221231', 
                train_range=12,
                validate_range=6,
                test_range=6,
                save=True,
                backtest=True)
```

```python
lgbm.train()
```

```python
lgbm.performance
```

```python
lgbm.plot_nav()
```

```python
lgbm.plot_longshort()
```
